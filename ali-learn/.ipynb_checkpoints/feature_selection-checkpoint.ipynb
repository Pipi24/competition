```python
# VarianceThreshold
from sklearn.feature_selection import VarianceThreshold
from sklearn.datasets import load_iris

iris = load_iris()
# 方差选择法，返回为特征选择后的数据
VarianceThreshold(threshold=3).fit_transform(iris.data)
```

```python
# SelectKBest
import numpy as np
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from array import array

# 1. 皮尔逊相关系数
from scipy.stats import pearsonr

iris = load_iris()
# map() 会根据提供的函数对指定序列做映射。
# 第一个参数 function 以参数序列中的每一个元素调用 function 函数，返回包含每次 function 函数返回值的新列表。
# 只能计算两组数据之间的皮尔森相关系数，不能批量计算。
SelectKBest(lambda X,Y : np.array(list(map(lambda x: pearsonr(x,Y), X.T))).T[0], 
            k=2).fit_transform(iris.data, iris.target)
```

```python
# 2.卡方检验
# 卡方检验是检验定性自变量和定性因变量的相关性，样本频数和期望频数的差距
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

iris = load_iris()
SelectKBest(chi2, k=2).fit_transform(iris.data, iris.target)
```

```python
# RFE
# 递归特征消除法
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# estimator：基模型
# n_features_to_select：选择的特征个数
# LogisticRegression： https://www.cnblogs.com/wjq-Law/p/9779657.html
RFE(estimator=LogisticRegression(multi_class='auto',
                                solver='lbfgs',
                                max_iter=500),
   n_features_to_select=2).fit_transform(iris.data, iris.target)
```

```python
# SelectFromModel
from sklearn.feature_selection import SelectFromModel
# 基于逻辑回归模型
from sklearn.linear_model import LogisticRegression

SelectFromModel(LogisticRegression(penalty='l2',
                                  C=0.1,
                                  solver='lbfgs',
                                  multi_class='auto')).fit_transform(iris.data, iris.target)
```

```python
# SelectFromModel
from sklearn.feature_selection import SelectFromModel
# 基于GBDT
from sklearn.ensemble import GradientBoostingClassifier

SelectFromModel(GradientBoostingClassifier()).fit_transform(iris.data, iris.target)
```
