### 评估指标
本赛题使用AUC（Area Under Curve）作为赛题的评估指标，其包含两方面的含义
- AUC反映的是模型对正负样本排序能力的强弱，对正负类分类的得分精度没有要求
- AUC越高，ROC曲线越靠近（0，1）点，模型排序能力越强

### 解题思路
一般来说，解题思路可以总结如下：
1. 赛题理解（理解赛题背景、评估指标，数据字段意义等）
2. EDA（缺失值，异常值，相关性，数据分布）
3. 特征工程（专业知识理解，特征选择、降维、生成、优化）
4. 模型训练（分类or回归，模型选择，参数调优）
5. 模型融合（vote, bagging，boosting，Averaging，ranking，blending，stacking）
